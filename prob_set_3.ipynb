```python
#Problem Set 3, Question 1:
import csv

def find_highest_water_level(file_path):
    highest_level = float('-inf')
    highest_time = ""

    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            water_level = float(row['Preliminary (ft)'])
            if water_level > highest_level:
                highest_level = water_level
                highest_time = f"{row['Date']} {row['Time (GMT)']}"

    return highest_level, highest_time

file_path = '/blue/bsc4452/share/Class_Files/data/CO-OPS_8727520_wl.csv'
highest_level, highest_time = find_highest_water_level(file_path)
print(f"Highest water level: {highest_level} ft at {highest_time}")
```

```python
#Problem Set 3, Question 2:
def calculate_water_levels(file_path):
    highest_level = float('-inf')
    lowest_level = float('inf')
    total_level = 0
    count = 0
    highest_time = ""
    lowest_time = ""

    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            water_level = float(row['Preliminary (ft)'])
            total_level += water_level
            count += 1
            if water_level > highest_level:
                highest_level = water_level
                highest_time = f"{row['Date']} {row['Time (GMT)']}"
            if water_level < lowest_level:
                lowest_level = water_level
                lowest_time = f"{row['Date']} {row['Time (GMT)']}"

    average_level = total_level / count
    return lowest_level, lowest_time, highest_level, highest_time, average_level

lowest_level, lowest_time, highest_level, highest_time, average_level = calculate_water_levels(file_path)
print(f"Lowest water level: {lowest_level} ft at {lowest_time}")
print(f"Highest water level: {highest_level} ft at {highest_time}")
print(f"Average water level: {average_level} ft")
```

```python
#Problem Set 3, Question 3:
def fastest_rise_in_water_level(file_path):
    previous_level = None
    fastest_rise = float('-inf')
    fastest_time = ""

    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            current_level = float(row['Preliminary (ft)'])
            if previous_level is not None:
                rise = current_level - previous_level
                if rise > fastest_rise:
                    fastest_rise = rise
                    fastest_time = f"{row['Date']} {row['Time (GMT)']}"
            previous_level = current_level

    return fastest_rise, fastest_time

fastest_rise, fastest_time = fastest_rise_in_water_level(file_path)
print(f"Fastest rise in water level: {fastest_rise} ft at {fastest_time}")
```

```python
#Problem Set 3, Question 3:
def monitor_water_levels(file_path):
    previous_level = None

    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row['Preliminary (ft)'] == '':
                print(f"No reading received at {row['Date']} {row['Time (GMT)']}")
                continue

            current_level = float(row['Preliminary (ft)'])
            if previous_level is not None:
                if current_level - previous_level > 0.25:
                    print(f"Warning: Water level increased by more than 0.25 ft at {row['Date']} {row['Time (GMT)']}")
            if current_level > 5.0:
                print(f"Warning: Water level over 5.0 ft at {row['Date']} {row['Time (GMT)']}")
            previous_level = current_level

monitor_water_levels(file_path)
```
